```python
########## Atividade 5 ##########
lista_vendas = [1000, 500, 800, 1500, 2000, 2300]
meta = 1200
percetual_bonus = 0.1
bonus_extra = 50

for venda in lista_vendas:
  bonus = 0

  if venda >= meta:
    bonus = venda * percetual_bonus

    if venda >= 2000:
      bonus = bonus + bonus_extra

  print(f"Venda: {venda} -> Bonus: R${bonus}")
```

```python
########## Atividade 6 ##########
lista_vendas = [1000, 500, 800, 1500, 2000, 2300]

maior_venda = lista_vendas[0]

for venda in lista_vendas:
  if venda > maior_venda:
    maior_venda = venda

print(maior_venda)
```
